```python
poke_groups = pokemons.groupby(by=['Type 1', 'Type 2'], dropna=False)
poke_groups.groups
```

```python
pokemons.groupby('Type 1').agg({'Total': ['min','max','mean']}).round(2)
```

```python
# works with multiple columns
pk_grp = pokemons.groupby(by=['Type 1', 'Type 2'], dropna=False)
pk_grp.agg({'Total': ['min','max','mean']}).round(2)
```

```python
exams = pd.read_csv('exams.csv')
exams
```

```python
Total = exams['math score'] + exams['reading score'] + exams['writing score']
Total
exams['Total'] = Total
exams
```

```python
exam_grp =exams.groupby(by=['gender', 'lunch'], dropna=False)
exam_grp.groups
```

```python
stats = exam_grp.agg({'Total': ['min','max','mean']}).round(2)
stats
```

```python
#stats.sort_values(by=('Total', 'mean'), ascending=False) both are same
stats['Total'].sort_values(by='mean', ascending=False)
```
