```python
def strength(attack: int) -> str:
    if attack > 150:
        return 'Strong'
    elif attack > 100:
        return 'Average'
    else:
        return 'Weak'
    
```

```python
def score(row):
    attack: int = row['Attack']
    speed: int = row['Speed']
    defense: int = row['Defense']
    return sum([attack, speed, defense])
```

```python
pokemons['Score'] = pokemons.apply(score, axis=1)
pokemons
```
