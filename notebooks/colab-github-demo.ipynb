```
df.assign(ratio=df["total_bedrooms"]/df["population"]).head()
```

`assign()` - always returns a copy of the data, leaving the original DataFrame untouched

```
(df.query("housing_median_age > 35").assign(bed_ratio=df["total_bedrooms"]/df["population"],
                                            room_ratio=df["total_rooms"]/df["population"])
.plot(kind="scatter", x="bed_ratio", y="room_ratio")
)
```
